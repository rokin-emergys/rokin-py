#### Question 1. Load the (states.json) data from the JSON file into a Python data structure.

```python
import json
```

```python
# 1. Load the states.json data
with open('states.json', 'r') as file:
    states_data = json.load(file)
states_data
```

```python
# 2. Extract the names of all states with multiple area codes
states = states_data['states']
states_with_multiple_area_codes = [
    state['name'] for state in states if len(state['area_codes']) > 1
]
print(states_with_multiple_area_codes)
```

```python
# 3. Calculate the total number of area codes for each state and store in a dictionary
state_area_code_count = {
    state['name']: len(state['area_codes']) for state in states

}
print(state_area_code_count)
```

```python
# 4. Identify the state with the highest number of area codes
state_with_max_area_codes = max(state_area_code_count, key=state_area_code_count.get)
max_area_codes = state_area_code_count[state_with_max_area_codes]
```

```python
print(f"States with multiple area codes: {states_with_multiple_area_codes}\n")
print(f"Total number of area codes per state: {state_area_code_count}\n")
print(f"\nState with the highest number of area codes: {state_with_max_area_codes} ({max_area_codes} area codes)")
```

```python
# 5. Convert Python dictionary to JSON (sorted by key) and print with indent level 4
sorted_json_data = json.dumps(state_area_code_count, indent=4, sort_keys=True)
print("\nSorted JSON data:")
print(sorted_json_data)
```
