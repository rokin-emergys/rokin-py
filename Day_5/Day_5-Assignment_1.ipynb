```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import re


driver = webdriver.Chrome()

url = "https://www.accuweather.com/en/in/mumbai/204842/weather-forecast/204842"
driver.get(url)


time.sleep(5)
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')

# Extract weather information
forecast_items = soup.find_all('a', class_='daily-list-item')
forecast_items

for item in forecast_items:
    # Etract day
    day = item.find('p', class_='day').text

    # Extract high and low temperatures 
    high_temp = item.find('span', class_='temp-hi').text
    low_temp = item.find('span', class_='temp-lo').text

    # Extract the weather phrase
    weather_phrase = item.find('p', class_='no-wrap').text


    # Print the extracted details
    print(f"Day: {day}")
    print(f"High Temperature: {high_temp}")
    print(f"Low Temperature: {low_temp}")
    print(f"Weather: {weather_phrase}")
    print("-" * 40)
driver.quit()
```

```python
import random
import threading


coffee_types = ["Espresso", "Latte", "Cappuccino", "Americano"]
sizes = ["Small", "Medium", "Large"]
milk_types = ["Whole", "Skim", "Oat", "Almond"]

def place_order(customer_id):
    """Simulates placing an order and processing it with a random prep time."""
    coffee = random.choice(coffee_types)
    size = random.choice(sizes)
    milk = random.choice(milk_types)
    prep_time = random.randint(5, 18)

    print("=" * 50)
    print(f"Customer {customer_id} ordered a {size} {coffee} with {milk} milk.")
    print(f"Estimated wait time for Customer {customer_id}: {prep_time} minutes.")

    time.sleep(prep_time)
    print("*" * 50)
    print(f"Customer {customer_id}'s {size} {coffee} is ready! (Took {prep_time} minutes) ✅\n")


num_customers = 5  # Number of customers
threads = []

for i in range(1, num_customers + 1):
    thread = threading.Thread(target=place_order, args=(i,))
    threads.append(thread)
    thread.start()

print()
print('#' * 50)
print()

for thread in threads:
    thread.join()

print("All orders have been processed! ✅")
```

```python
import requests
import multiprocessing
import os

session = requests.Session()

def download_image(image_id):
    """Download an image and save it to a dedicated directory using a session"""
    directory = f"images/image_{image_id}"
    
    os.makedirs(directory, exist_ok=True)

    url = f"https://picsum.photos/2000/3000?random={image_id}"
    response = session.get(url, stream=True)
    
    filename = os.path.join(directory, f"image_{image_id}.jpg")
    with open(filename, "wb") as f:
        for chunk in response.iter_content(1024):
            f.write(chunk)
    
    print(f"Downloaded image_{image_id}.jpg to {directory}")


# Without multiprocessing
def download_images_sequential():
    """Download images sequentially"""
    start_time = time.time()
    for i in range(1, 6):
        download_image(i)
    
    end_time = time.time()
    print(f"\nSequential download time: {end_time - start_time:.2f} seconds")
```

```python
def download_images_multiprocessing():
    """Download images using multiprocessing"""
    start_time = time.time()
    
    with multiprocessing.Pool(processes=min(10, multiprocessing.cpu_count())) as pool:
        pool.map(download_image, range(1, 6))  # Download images 1-5 in parallel
    
    end_time = time.time()
    print(f"\nMultiprocessing download time: {end_time - start_time:.2f} seconds")
```

```python
# Without Multiprocessing
download_images_sequential()
```

```python
download_images_multiprocessing()
```
