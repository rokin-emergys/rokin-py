```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import re
# Set up the Selenium WebDriver (make sure to specify the path to your ChromeDriver)
# service = Service("path_to_chromedriver")  
driver = webdriver.Chrome()

# Open the website
url = "https://www.accuweather.com/en/in/mumbai/204842/weather-forecast/204842"
driver.get(url)

# Allow page to load
time.sleep(5)

# Get page source
page_source = driver.page_source

# Use BeautifulSoup to parse the HTML
soup = BeautifulSoup(page_source, 'html.parser')

# Find the forecast container
forecast_card = soup.find('div', class_='today-forecast-card content-module')

if forecast_card:
    # Extract the day
    day = forecast_card.find('div', class_='card-header spaced-content').find('p').text.strip()
    
    # Extract high and low temperatures
    body_items = forecast_card.find('div', class_='body').find_all('div', class_='body-item')
    
    high_temp_raw = body_items[0].find('p').text.strip()
    low_temp_raw = body_items[1].find('p').text.strip()

    high_temp = re.search(r'(\d+°)', high_temp_raw)
    low_temp = re.search(r'(\d+°)', low_temp_raw)

    high_temp = high_temp.group(1) if high_temp else "N/A"
    low_temp = low_temp.group(1) if low_temp else "N/A"
    print(f"Day: {day}, High: {high_temp}, Low: {low_temp}")

else:
    print("Weather data not found!")

# Close the driver
driver.quit()
```

```python
import threading
import time
import random

# List of coffee types and options
coffee_types = ["Espresso", "Latte", "Cappuccino", "Americano"]
sizes = ["Small", "Medium", "Large"]
milk_types = ["Whole", "Skim", "Oat", "Almond"]

def place_order(customer_id):
    """Simulates placing an order and processing it with a random prep time."""
    coffee = random.choice(coffee_types)
    size = random.choice(sizes)
    milk = random.choice(milk_types)
    prep_time = random.randint(5, 18)

    print("=" * 50)
    print(f"Customer {customer_id} ordered a {size} {coffee} with {milk} milk.")
    print(f"Estimated wait time for Customer {customer_id}: {prep_time} minutes.")

    time.sleep(prep_time)
    print("*" * 50)
    print(f"Customer {customer_id}'s {size} {coffee} is ready! (Took {prep_time} minutes) ✅\n")

# Creating and starting multiple customer threads
num_customers = 5  # Number of customers
threads = []

for i in range(1, num_customers + 1):
    thread = threading.Thread(target=place_order, args=(i,))
    threads.append(thread)
    thread.start()

print()
print('#' * 50)
print()

for thread in threads:
    thread.join()

print("All orders have been processed! ✅")
```

```python
import requests
import time
import multiprocessing
import os

def download_image(image_id):
    # Create a unique directory for each image
    directory = f"images/image_{image_id}"
    if not os.path.exists(directory):
        os.makedirs(directory)
        
    url = f"https://picsum.photos/2000/3000?random={image_id}"
    img_data = requests.get(url).content
    image_path = os.path.join(directory, f"image_{image_id}.jpg")
    
    with open(image_path, "wb") as f:
        f.write(img_data)
    print(f"Downloaded image_{image_id}.jpg to {directory}")

# Without multiprocessing
start_time = time.time()
for i in range(1, 6):
    download_image(i)
end_time = time.time()
print(f"Time without multiprocessing: {end_time - start_time} seconds")

# With multiprocessing
start_time = time.time()
with multiprocessing.Pool(processes=5) as pool:
    pool.map(download_image, range(1, 6))
end_time = time.time()
print(f"Time with multiprocessing: {end_time - start_time} seconds")
```
